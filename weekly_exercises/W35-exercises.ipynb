### Derivative 1:
1. Given: $$
 \frac{\partial (\boldsymbol{b}^T\boldsymbol{a})}{\partial \boldsymbol{a}} = \boldsymbol{b}.
 $$
2. Scalar Product \
 $ f(a) = \sum_{i=1}^{n} b_i a_i\ $, where $n$ is the dimension of the vectors.

3. Differentiation \
 $ \frac{{\partial f}}{{\partial a}} = \frac{{\partial}}{{\partial a}} \left( \sum_{i=1}^{n} b_i a_i \right) $ 

4. Chain rule  
 $ \frac{{\partial f}}{{\partial a}} = \sum_{i=1}^{n} b_i \frac{{\partial a_i}}{{\partial a}} $ 

5. Derivative of $ a_i $\
Since $ a_i $ is just the $ i$-th element of vector $ \boldsymbol{a} $, its derivative with respect to $ a $ is $1$ for the $ i$-th element and $0$ for others.
6. Substitute and Simplify\
$\frac{{\partial f}}{{\partial a}} = \sum_{i=1}^{n} b_i \cdot 1 \cdot e_i$, where $e_i$ is the $i$-th standard basis vector.\
\
$\frac{{\partial f}}{{\partial a}} = \sum_{i=1}^{n} b_i e_i = b$, where $b$ is the vector and $b_i$ are its elements.

So, $\frac{{\partial (b^T a)}}{{\partial a}} = b$.
\
This demonstrates that the derivative of $b^T a$ with respect to $a$ is indeed $b$.

### Derivative 2:

1. Given: $$
\frac{\partial (\boldsymbol{a}^T\boldsymbol{A}\boldsymbol{a})}{\partial \boldsymbol{a}} = \boldsymbol{a}^T(\boldsymbol{A}+\boldsymbol{A}^T)
$$
2. $ f(a) = a^TAa$

3. $ f(a) = a^TAAa$

4. Differentiation: \
$ \frac{\partial f}{\partial a} = \frac{\partial}{\partial a}(a^TAAa) $

5. Chain rule: \
$ \frac{\partial f}{\partial a} = \frac{\partial}{\partial a}(a^T) \cdot (AAa) + a^T \cdot \frac{\partial}{\partial a}(AAa) $

6. Derivative of $ a^T $ \
The derivative of $a^T$ with respect to $a$ is the matrix $I$ (identity matrix), because it's like taking the derivative of a constant with respect to $a$.

7. Derivative of $AAa$ with respect to $a$ (again I'm applying the chain rule) \
$\frac{\partial}{\partial a}(AAa) = A \cdot \frac{\partial}{\partial a}(Aa)$

8. Derivative of $Aa$ with respect to $a$ \
To find the derivative of $Aa$ with respect to $a$, we have just $A$ because it's a linear transformation.

9. $ \frac{\partial (a^TAa)}{\partial a} = I \cdot (AAa) + a^T \cdot (A \cdot A) = AAa + a^T(AA) $

10. Transpose of $A$ \
Since $A$ is a matrix: $A^T = A$.
\
\
\
\
Substitute $A^T = A$ into the expression:\
$\frac{\partial (a^TAa)}{\partial a} = AAa + a^T(AA) = AAa + a^T(AA^T) = AAa + a^TA^TA = a^T(A + A^T)$ \
\
So, $\frac{\partial (a^TAa)}{\partial a} = a^T(A + A^T)$.

This demonstrates the derivative of $a^TAa$ with respect to $a$.

### Derivative 3:

1. Given: $$
\frac{\partial \left(\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s}\right)^T\left(\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s}\right)}{\partial \boldsymbol{s}} = -2\left(\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s}\right)^T\boldsymbol{A}
$$

2. $f(s) = (x - As)^T(x - As)$

3. $f(s) = (x - As)^T(x - As)$

4. Differentiation:\
$\frac{\partial f}{\partial s} = \frac{\partial}{\partial s}((x - As)^T(x - As))$

5. Chain rule:\
$\frac{\partial f}{\partial s} = \frac{\partial}{\partial s}(x - As)^T \cdot (x - As) + (x - As)^T \cdot \frac{\partial}{\partial s}(x - As)$

6. Derivative of $x - As$ with respect to $s$\
To find the derivative of $x - As$ with respect to $s$, we have just a $-A$ because it's a linear transformation.

7. $\frac{\partial ((x - As)^T(x - As))}{\partial s} = -A^T(x - As) - (x - As)^TA $

8. Distribution of $-1$ \
$\frac{\partial ((x - As)^T(x - As))}{\partial s} = -(A^T(x - As) + (x - As)^TA)$

9. Combination of the two terms: \
$\frac{\partial ((x - As)^T(x - As))}{\partial s} = -(A^T(x - As) + (x - As)^TA) = -((x - As)^TA + (x - As)^TA)$

10. Factor out $-2$ \
$\frac{\partial ((x - As)^T(x - As))}{\partial s} = -2((x - As)^TA) $
\
\
\
\
So, $\frac{\partial ((x - As)^T(x - As))}{\partial s} = -2((x - As)^TA)$.

This demonstrates the derivative of $(x - As)^T(x - As)$ with respect to $s$.


## Exercise 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


x = np.random.rand(100, 1)
y = 2.0 + 5 * x * x + 0.1 * np.random.randn(100, 1)

X = np.column_stack((np.ones_like(x), x, x**4))

linreg = LinearRegression()
linreg.fit(X, y)

y_predict = linreg.predict(X)

print('Intercept: ', linreg.intercept_)
print('Coefficients: ', linreg.coef_)
print('MSE: %.2f' % mean_squared_error(y, y_predict))
print('R-squared: %.2f' % r2_score(y, y_predict))

plt.scatter(x, y, marker='o')
plt.plot(x, y_predict, "ro", label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()
```

## Exercise 3

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Setting a random seed, data points and max. polynomial degree to test
np.random.seed(123)
n = 100
maxdegree = 15

# Generating the data
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2) + np.random.normal(0, 0.1, x.shape)

# Arrays to store errors and polynomial degrees
TestError = np.zeros(maxdegree)
TrainError = np.zeros(maxdegree)
polydegree = np.zeros(maxdegree)

# Splitting data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

# Standardizing the features using StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Loop over different polynomial degrees
for degree in range(maxdegree):
    # Creating a pipeline for polynomial regression
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression(fit_intercept=False))

    # Fitting the model to the training data
    clf = model.fit(x_train_scaled, y_train)

    # Predicting on training and test data
    y_fit = clf.predict(x_train_scaled)
    y_pred = clf.predict(x_test_scaled)

    # Storing the current polynomial degree and errors
    polydegree[degree] = degree
    TestError[degree] = np.mean(np.mean((y_test - y_pred)**2))
    TrainError[degree] = np.mean(np.mean((y_train - y_fit)**2))

# Plotting the training and test errors
plt.plot(polydegree, TestError, label = 'Test Error', color='green')
plt.plot(polydegree, TrainError, label = 'Train Error', color='violet')
plt.legend()
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error')
plt.title('Polynomial Regression')
plt.show()
```
